```
df_silver_agg=(spark.readStream.format("delta").table("VehicleSensor.VehicleDelta_Silver"))
df_silver_agg.createOrReplaceTempView("vw_AggDetails")
```

```
%sql
SELECT Make,Model,coalesce(Borough,"UnKnown")AS Borough,lfi AS LowFuelIndicator,Location,Year,Month,Day,COUNT(*) as TotalVehicles
FROM vw_AggDetails
WHERE lfi=1
GROUP BY Year,Make,Model,coalesce(Borough,"UnKnown"),Location,Month,Day,lfi
ORDER BY Year DESC,Month DESC,TotalVehicles DESC
```

```
%sql
SELECT Id as VehicleId,Make,Model,coalesce(Borough,"Unknown") AS Borough,Location,Year,Month,Day,COUNT(*) as TotalVehicles
FROM vw_AggDetails
GROUP BY Make,Model,coalesce(Borough,"Unknown"),Location,Year,Month,Day,Id
```

```
%sql
SELECT Make,Borough,Year,Month,COUNT(*) as TotalVehicles
FROM vw_AggDetails
GROUP BY Make,Borough,Year,Month
```
